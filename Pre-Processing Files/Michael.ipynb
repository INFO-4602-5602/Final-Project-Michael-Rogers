# Access

## All of these calculations were done in Amazon Sagemaker. The GainLoss.txt is the file we will be visualizing. The calculations were done using python's CNVKit library.

```python
%%time
!cnvkit.py access hg19.fa -o michael_access.hg19.bed
```

```python
%%time
!cnvkit.py autobin AW85Y-30360-3-MCT2_S11.bam -t phil.bed -g michael_access.hg19.bed --annotate refFlat.txt --short-names
```

```python
%%time
!cnvkit.py coverage AW85Y-30360-3-MCT2_S11.bam phil.target.bed -o michael.targetcoverage.cnn
```

```python
%%time
!cnvkit.py coverage AW85Y-30360-3-MCT2_S11.bam phil.antitarget.bed -o michael.antitargetcoverage.cnn
```

```python
%%time
!cnvkit.py coverage AW85Y-30360-3-MCT1_S10.bam phil.target.bed -o michael_normal.targetcoverage.cnn
```

```python
%%time
!cnvkit.py coverage AW85Y-30360-3-MCT1_S10.bam phil.antitarget.bed -o michael_normal.antitargetcoverage.cnn
```

```python
%%time
!cnvkit.py reference normal.{,anti}targetcoverage.cnn --fasta hg19.fa -o michael_reference.cnn
```

```python
%%time
!cnvkit.py fix michael.targetcoverage.cnn michael.antitargetcoverage.cnn michael_reference.cnn -o michael.cnr
```

```python
%%time
!cnvkit.py segment michael.cnr -m haar -o michael.cns
```

```python
%%time
!cnvkit.py genemetrics michael.cnr -s michael.cns -t 0.4 -m 5 -y -o michael_gainloss.txt
```
